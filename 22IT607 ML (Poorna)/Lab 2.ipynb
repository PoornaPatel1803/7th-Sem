```python
#TRANSPOSE
#it is used in matrix multiplication it 
arr = np.array([[1, 2, 3, 4, 5, 6],
                [4, 5, 6 ,9, 6, 12]])

print(np.transpose(arr))
```

# Staring of matplotLib

```python
#line chart

from matplotlib import pyplot as plt
x = np.arange(15)
y = np.random.randint(low=0, high=100, size=15)
print("X::",x,"\n\nY::",y)
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(15)
y = np.random.randint(low=0, high=100, size=15)
print("X::",x,"\n\nY::",y)
plt.plot(x,y,linewidth='5')
plt.show()
```

```python
x = np.arange(15)
y = np.random.randint(low=0, high=100, size=15)
print("X::",x,"\n\nY::",y)
plt.plot(x,y,linewidth='5',color = 'orange')
plt.show()
```

```python
x1 = np.arange(15)
y1 = np.random.randint(low=0, high=100, size=15)
x2 = np.random.randint(low=0, high=100, size=15)
y2 = np.random.randint(low=0, high=100, size=15)
plt.xlabel("<-X->")
plt.ylabel("<-Y->")
print("X1::",x1,"\n\nY1::",y1,"\n\nX2::",x2,"\n\nY2::",y2)
plt.plot(np.sort(x1),np.sort(y1),np.sort(x2),np.sort(y2))
plt.show()
```

```python
#bar chart
course = np.array(["AI", "ML","PYTHON","PROJECT","PM"])
enroll = np.random.randint(low=0,high=60 ,size=5)
plt.xlabel("Course of institue")
plt.ylabel("Enrollment in course")
plt.title("course vs Enrollment")
plt.bar(course, enroll,width = 0.3 ,color="yellow", label="BVM Engineering", edgecolor="red")

course2 = np.array(["AI", "ML","PYTHON","PROJECT","PM"])
enroll2 = np.random.randint(low=0,high=60, size=5)
plt.bar(course2, enroll2,width = 0.3 ,color="green", label="DDU Engineering", edgecolor="blue")

plt.legend()
plt.show()
```

```python
#HISTOGRAM
data = np.random.randn(1000)

plt.hist(data, bins=10, color='coral', edgecolor='skyblue')

plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Basic Histogram')

plt.show()
```

```python
#SCATTER PLOT
import matplotlib.pyplot as plt


x = np.random.randint(low=10,high=10000,size=100)
y = np.random.randint(low=10,high=10000,size=100)
plt.xlabel("<-X->")
plt.ylabel("<-Y->")

plt.scatter(x,y,color="yellow",label="blue")

x = np.random.randint(low=10,high=10000,size=100)
y = np.random.randint(low=10,high=10000,size=100)

plt.scatter(x,y,color="green",label="red")
plt.legend()

plt.show()
```

```python
x = np.random.randint(low=10,high=10000,size=10)

plt.pie(x)
plt.show()
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["C++", "C", "JAVA", "PYTHON"]
plt.legend()
plt.pie(y, labels = mylabels)
plt.show() 
```
